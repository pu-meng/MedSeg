CE Loss

```python
import torch
import torch.nn.functional as F

# ========== 多分类 CE Loss ==========
# 模型输出
logits = torch.randn(2, 3, 4, 4, 4)  # [B=2, C=3, D=4, H=4, W=4]
target = torch.randint(0, 3, (2, 4, 4, 4))  # [B=2, D=4, H=4, W=4]

# 计算loss
loss = F.cross_entropy(logits, target)
print(f"CE Loss: {loss.item():.4f}")

# 手动计算（理解原理）
probs = F.softmax(logits, dim=1)  # [2, 3, 4, 4, 4]

# 单个体素的例子
b, d, h, w = 0, 2, 1, 3
true_class = target[b, d, h, w].item()  # 假设是2
pred_prob = probs[b, true_class, d, h, w].item()
manual_loss = -torch.log(torch.tensor(pred_prob))
print(f"单个体素的loss: {manual_loss.item():.4f}")


# ========== 二分类 CE Loss ==========
logits_binary = torch.randn(2, 1, 4, 4, 4)  # [B=2, C=1, D=4, H=4, W=4]
target_binary = torch.randint(0, 2, (2, 1, 4, 4, 4)).float()

probs_binary = torch.sigmoid(logits_binary)
loss_binary = F.binary_cross_entropy(probs_binary, target_binary)
print(f"Binary CE Loss: {loss_binary.item():.4f}")

# 手动计算
b, c, d, h, w = 0, 0, 2, 1, 3
y = target_binary[b, c, d, h, w].item()
p = probs_binary[b, c, d, h, w].item()
p=torch.tensor(p)
manual_binary_loss = -(y * torch.log(torch.tensor(p)) + (1-y) * torch.log(1-p))
print(f"单个体素的binary loss: {manual_binary_loss.item():.4f}")
```
